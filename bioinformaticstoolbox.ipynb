```python

DNANucleotides = ['A', 'C','G','T']
ReverseComp = {"A":"T", "T":"A", "C":"G", "G":"C"}
dna_seq = input("Type your DNA Strand here:") # You input your DNA strand
DNAString = "TTTCACAGCGATCGCGCGTTTTACACTAC" # Sample DNA String
# Function 1: Validating a sequence if its true ------------------------------------
def ValidateSeq(dna_seq):
  tmpseq = dna_seq.upper()
  for nuc in tmpseq:
      if nuc not in DNANucleotides:
        return print("Sorry that's not a DNA Strand :(")
        break
  return tmpseq
print("[1] Yay, this is your DNA Strand:" + ValidateSeq(dna_seq)) #

# Function 2 Random DNA Generator --------------------------------------------------
import random
randDNAstr = ''.join([random.choice(DNANucleotides) for nuc in range(20)])
print("[2] Here is a randomly generated DNA: " + ValidateSeq(randDNAstr))
DNAstring = ValidateSeq(randDNAstr)

# Function 3 Counting Nucleotides in DNA ---------------------------------------------------
def countNucfrequencyDNA(seq):
  temporaryFrequencyDictionary = {"A": 0, "C": 0, "G": 0, "T": 0}
  for nuc in DNAstring:
    temporaryFrequencyDictionary[nuc] += 1
  return temporaryFrequencyDictionary
print("[3] DNA Strand:" + ValidateSeq(randDNAstr))
print(countNucfrequencyDNA(DNAstring))
print("[4] Nucleotide Frequency Length:", {len(DNAstring)})

# Function 4 Transcription DNA -> RNA ------------------------------------------------------
def transcription(seq):
  # Replaces Thymine with Uracil in DNA
  return seq.replace("T", "U")
print("[5] DNA/RNA Transcription: " + transcription(DNAstring)) # Turned DNA -> RNA
RNAstring = transcription(DNAstring)

# Function 5 Reverse Complementary DNA Strand --------------------------------
def reversecomplement(seq):
  mapping = {"A":"T", "T":"A", "C":"G", "G":"C"}
  return "".join([mapping[base] for base in seq[::-1]])
print("[6] This is our complementary strand:", reversecomplement(DNAstring))


# Putting Reverse Complementary and Original STrand together -------
print("[7] Making a Actual DNA strand with Reverse Comple + DNA: ")
print("5' " + (DNAstring) + " 3'")
print( '   ' + ''.join(['|' for c in range(len(DNAstring))]))
print("3' " + reversecomplement(DNAstring) + " 5' ")

# Function 6 Counting Nucleotides in RNA --------------------------------------
def countNucfrequencyRNA(seq):
  temporaryFrequencyDictionary = {"A": 0, "C": 0, "G": 0, "U": 0}
  for nuc in RNAstring:
    temporaryFrequencyDictionary[nuc] += 1
  return temporaryFrequencyDictionary
print("[8] RNA Strand:" + RNAstring)
print(countNucfrequencyRNA(RNAstring))


# Function 7 Colorizing DNA Ouput for Presentation (Python Preference) ---------------------------
def colored_montified(seq):
  bcolors = {
      'A': '\033 [92n',
      'C': '\033 [94n',
      'G': '\033 [93n',
      'T': '\033 [91n',
      'U': '\033 [91n',
      'reset': '\033 [0;0n'
  }
  tmpStr = ' '
  for nuc in seq:
    if nuc  in bcolors:
      tmpStr += bcolors[nuc] + nuc
    else:
      tmpStr += bcolors['reset'] + nuc
  return tmpStr
print("[9]This is our DNA with Colors" + colored_montified(DNAstring))
# This will only show up in the terminal code, You write "python filename.py"


# Function 8 GC Content Calculations ----------------------------------------
def gc_content (seq):
    return round ((seq.count("C") + seq.count("G")) / len(seq) * 100, 6)
print("[10] This is our GC content in DNA Strand:",gc_content(DNAstring), "%")

# Function 9 GC Content Calculation in Specific Window ------------------------
def gc_content_subseq(seq, k):
    res = [] # Our result stored here
    for i in range(0, len(seq) - k + 1, k):
        subseq = seq[i:i + k]
        res.append(gc_content(subseq))
    return res

print("[11] This is our GC content at k=5", gc_content_subseq(DNAstring, k=5))
#We can give K a value we want, its how much character we want to calculate our DNAstring by.
```

```python
from collections import Counter
#---------------------------- CODONS & Nucleotide DICTIONARIES: ---------------
# Chart with all DNA codons used for Translation
DNA_Codons = {
    # 'M' - START, '_' - STOP
    "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "TGT": "C", "TGC": "C",
    "GAT": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "TTT": "F", "TTC": "F",
    "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G",
    "CAT": "H", "CAC": "H",
    "ATA": "I", "ATT": "I", "ATC": "I",
    "AAA": "K", "AAG": "K",
    "TTA": "L", "TTG": "L", "CTT": "L", "CTC": "L", "CTA": "L", "CTG": "L",
    "ATG": "M",
    "AAT": "N", "AAC": "N",
    "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "CAA": "Q", "CAG": "Q",
    "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
    "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S", "AGT": "S", "AGC": "S",
    "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",
    "TGG": "W",
    "TAT": "Y", "TAC": "Y",
    "TAA": "_", "TAG": "_", "TGA": "_"
}

NUCLEOTIDE_BASE = {
    "DNA": ["A", "T","C", "G"],
    "RNA": ["A", "U","C", "G"],
}

RNA_Codons = {
    # 'M' - START, '_' - STOP
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "UGU": "C", "UGC": "C",
    "GAU": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "UUU": "F", "UUC": "F",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G",
    "CAU": "H", "CAC": "H",
    "AUA": "I", "AUU": "I", "AUC": "I",
    "AAA": "K", "AAG": "K",
    "UUA": "L", "UUG": "L", "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "AUG": "M",
    "AAU": "N", "AAC": "N",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "CAA": "Q", "CAG": "Q",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S", "AGU": "S", "AGC": "S",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "UGG": "W",
    "UAU": "Y", "UAC": "Y",
    "UAA": "_", "UAG": "_", "UGA": "_"
}



# [12] Translation into Amino Acids Function -------------------------------
def translate_seq(seq, init_pos=0):
    return [DNA_Codons[seq[pos:pos + 3]] for pos in range(init_pos, len(seq) - 2, 3)]
print(f'[12] Aminoacid Sequences from DNA: {translate_seq(DNAstring, 0)} \n')

# [13] Counts the Codons/Amino acid usage in specific DNA sequeunce --------------------
def codon_usage(seq, aminoacid):
    tmplist = []
    for i in range(0, len(seq) - 2, 3):
        if DNA_Codons[seq[i:i + 3]] == aminoacid:
            tmplist.append(seq[i:i + 3])
    freqDict = dict(Counter(tmplist))
    totalWight = sum(freqDict.values())
    for seq in freqDict:
        freqDict[seq] = round(freqDict[seq] / totalWight, 2)
    return freqDict
print(f'[13] This counts the amino acids of ("Specific Amino Acid"): {codon_usage(DNAstring, "V")}\n')
# You can change the letter depending on which amino acid you want to change _________________^



# [14] Generating Reading Frames for Proteins in FUture --------------------------------
def gen_reading_frames(seq):
    frames = []
    frames.append(translate_seq(seq, 0)) # Start from reading from start of frame
    frames.append(translate_seq(seq, 1)) # THis function works with 5 and
    frames.append(translate_seq(seq, 2)) # Three reading frames
    frames.append(translate_seq(reversecomplement(seq), 0))
    frames.append(translate_seq(reversecomplement(seq), 1))
    frames.append(translate_seq(reversecomplement(seq), 2))
    return frames
    #We have reversecomplement as well to complete the DNA

print("[14] This is our Reading Frames:")
for frame in gen_reading_frames(DNAstring):
    print(frame)


# [15] Protein Stop and Start Codon Analysis -------------------------------------
def protein_reading_analysis(aa_seq):
  # M is our start codon, _ is out stop codon according to our dictionary
  current_prot = []
  proteins = [] # <- Proteins added which are those after every stop codon
  for aa in aa_seq:
    if aa == "_": # <- Our current codon being searched
      # This Part searches for Start Codons
      if current_prot:
        for p in current_prot:
          proteins.append(p)
        current_prot = []
    else:
      # This Part searches for Stop Codons
      if aa == "M":
        current_prot.append("")
    for i in range(len(current_prot)):
      current_prot[i] += aa
  return proteins
test_amino_acids = ['T', 'M', 'H', 'I', 'K', 'T', 'L', 'H', 'G', 'L'
   'P', 'H', 'T', 'S', 'R', 'L', 'Y', 'T', '_']
print("\n[15] Here is our protein using Stop & Start Codons:")
print(protein_reading_analysis(test_amino_acids))


# [16] --- POINT MUTATIONS DETECTION ----------------------------------
# For this to work, we need two strings at same length, Example Below:
DNAString1 = "GAGCCTACTAACGGGAT"
DNAString2 = "CATCGTAATGACGGCCT"
# This function tracks how many mutations are there in the String
def hamming_distance_pointmutation (DNAString1, DNAString2):
  if len(DNAString1) != len(DNAString2):
    print("Strings are not equal size, you can't detect mutations")
  else:
    hamming_distance = 0
    for i in range(len(DNAString1)):
      if DNAString1[i] != DNAString2[i]:
        hamming_distance += 1
    return hamming_distance
print(f'[16]The amount of Point Mutations detected on two DNA strands are: {hamming_distance_pointmutation(DNAString1, DNAString2)}')


# [17] All Proteins in Open Reading Frames
def all_proteins_in_orfs (seq, startReadPos=0, endReadPos=0, ordered=False):
  # This function needs functions 12,14,and 15 to run
  if endReadPos > startReadPos:
    rfs = gen_reading_frames(seq[startReadPos:endReadPos])
  else:
    rfs = gen_reading_frames(seq)

  #Finds proteins
  res = []
  for rf in rfs:
    proteins = protein_reading_analysis(rf)
    for protein in proteins:
        res.append(protein)
  if ordered:
    return sorted(res, key=len, reverse=True)
  return res

  # Executes our function:
print(f'[17] All Proteins in ORFS Printed:')
for prot in all_proteins_in_orfs(DNAString, 0, 0, True):
  print(f"{prot}")
from os import write

# [18] READS A TEXTFILE ----------------------------------------------------
def readTextFile(filePath):
  with open(filepath, 'r') as f:
    return "".join([l.strip() for l in f.readlines()])
   # reads file, line by line and strips any new lines


# [19] WRITES A TEXTFILE --------------------------------------------------
def writeTextFile(filePath, seq, mode='w'):
  with open(filepath, mode) as f:
    f.write(seq + '\n')
    #Overwrites our file. Makes a file with output

#writeTextFile("test.txt, test_dna.seq")
#for rf in test_dna.gen_reading_frame():
  #writeTextFile("test.txt", str(rf), 'a')

# [20] READS A FASTAFILE -----------------------------------
def read_FASTA(filePath):
  with open(filePath, "r") as f:
    FASTAfile = [l.strip() for l in f.readlines()]
  FASTADic = {}
  FASTAlabel = ""
  for line in FASTAfile:
    if ">" in line:
      FASTAlabel = line[1:]
      FASTADic[FASTAlabel] = ""
    else:
      FASTADic[FASTAlabel] += line
  return FASTADic


```
